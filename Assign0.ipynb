### c) swap the red and green color channels of the input color image

```python
astronaut = (data.astronaut())

print(np.shape(astronaut))
red_astro = astronaut[:,:,0]
astronaut[:,:,0] = astronaut[:,:,1]
astronaut[:,:,1] = red_astro
plt.imshow((np.uint8(astronaut)))
plt.show()
```

#### d) average the input image with its mirror image

#### e) add or subtract a random value between [0,255] to every pixel in a grayscale image, then clip the resulting image to have a minimum value of 0 and a maximum value of 255.

```python
cat = data.chelsea()
cat = rgb2gray(cat)
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(10, 5))
#ax3 = plt.subplot(1, 3, 3)
ax0.imshow(cat, cmap='gray');
row, col = cat.shape
print ("row, col", row, col)
print("Shape:", cat.shape)
print("Values min/max:", cat.min(), cat.max())
randValues = (np.uint8(np.random.randn(row, col) *100))
cat = np.uint8(np.add(cat, randValues))
min_mask = cat < 0
max_mask = cat > 255
cat[min_mask] = 0
cat[max_mask] = 255
#print (cat)
#print ("Mean ", cat.mean())
ax1.imshow(cat, cmap='gray');
plt.show()
```
